```python
from IPython.display import HTML

html_content = '''
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Type-II Heterojunction Performance Prediction Tool for Photodynamic Materials</title>
    <script src="https://cdn.jsdelivr.net/npm/chart.js"></script>
    <script src="https://cdn.jsdelivr.net/npm/xlsx@0.18.5/dist/xlsx.full.min.js"></script>
    <link rel="stylesheet" href="https://cdnjs.cloudflare.com/ajax/libs/font-awesome/6.4.0/css/all.min.css">
    <style>
        :root {
            --primary: #3498db;
            --secondary: #2ecc71;
            --accent: #e74c3c;
            --light: #ecf0f1;
            --dark: #2c3e50;
            --darker: #1a252f;
            --shadow: 0 4px 15px rgba(0, 0, 0, 0.1);
            --gradient-start: #1e5799;
            --gradient-end: #207cca;
        }

        * {
            margin: 0;
            padding: 0;
            box-sizing: border-box;
            font-family: 'Segoe UI', Tahoma, Geneva, Verdana, sans-serif;
        }

        body {
            background: linear-gradient(135deg, #f5f7fa 0%, #e4edf5 100%);
            color: var(--dark);
            min-height: 100vh;
            padding: 20px;
            line-height: 1.6;
        }

        .container {
            max-width: 1200px;
            margin: 0 auto;
        }

        header {
            text-align: center;
            padding: 30px 0;
            margin-bottom: 30px;
            background: linear-gradient(120deg, var(--gradient-start), var(--gradient-end));
            border-radius: 15px;
            box-shadow: var(--shadow);
            color: white;
            position: relative;
            overflow: hidden;
        }

        header::before {
            content: "";
            position: absolute;
            top: 0;
            left: 0;
            width: 100%;
            height: 100%;
            background: url('data:image/svg+xml;utf8,<svg xmlns="http://www.w3.org/2000/svg" viewBox="0 0 200 200"><circle cx="50" cy="50" r="8" fill="rgba(255,255,255,0.1)"/><circle cx="150" cy="100" r="10" fill="rgba(255,255,255,0.1)"/><circle cx="80" cy="180" r="6" fill="rgba(255,255,255,0.1)"/></svg>');
        }

        h1 {
            font-size: 2.8rem;
            margin-bottom: 15px;
            text-shadow: 0 2px 4px rgba(0,0,0,0.2);
            position: relative;
        }

        .subtitle {
            font-size: 1.3rem;
            max-width: 800px;
            margin: 0 auto;
            opacity: 0.9;
            position: relative;
        }

        .main-content {
            display: grid;
            grid-template-columns: 1fr 1fr;
            gap: 30px;
            margin-bottom: 40px;
        }

        @media (max-width: 900px) {
            .main-content {
                grid-template-columns: 1fr;
            }
        }

        .card {
            background: white;
            border-radius: 15px;
            padding: 30px;
            box-shadow: var(--shadow);
            transition: transform 0.3s ease, box-shadow 0.3s ease;
        }

        .card:hover {
            transform: translateY(-10px);
            box-shadow: 0 10px 25px rgba(0, 0, 0, 0.15);
        }

        .section-title {
            font-size: 1.8rem;
            margin-bottom: 25px;
            padding-bottom: 15px;
            border-bottom: 3px solid var(--primary);
            color: var(--darker);
            display: flex;
            align-items: center;
        }

        .section-title i {
            margin-right: 15px;
            background: var(--primary);
            width: 40px;
            height: 40px;
            border-radius: 50%;
            display: flex;
            align-items: center;
            justify-content: center;
            color: white;
        }

        .input-group {
            margin-bottom: 25px;
            position: relative;
        }

        label {
            display: block;
            margin-bottom: 12px;
            font-weight: 600;
            color: var(--darker);
            font-size: 1.1rem;
        }

        input, select {
            width: 100%;
            padding: 15px;
            border: 2px solid #ddd;
            border-radius: 10px;
            font-size: 1.1rem;
            transition: all 0.3s;
            background: #fafafa;
        }

        input:focus, select:focus {
            border-color: var(--primary);
            outline: none;
            background: white;
            box-shadow: 0 0 0 3px rgba(52, 152, 219, 0.2);
        }

        .btn-group {
            display: flex;
            gap: 15px;
            margin-top: 30px;
            flex-wrap: wrap;
        }

        button {
            padding: 15px 25px;
            border: none;
            border-radius: 10px;
            font-size: 1.1rem;
            font-weight: 600;
            cursor: pointer;
            transition: all 0.3s ease;
            flex: 1;
            display: flex;
            align-items: center;
            justify-content: center;
            min-width: 150px;
        }

        button i {
            margin-right: 10px;
        }

        .btn-primary {
            background: linear-gradient(135deg, var(--primary), #2980b9);
            color: white;
        }

        .btn-secondary {
            background: linear-gradient(135deg, var(--secondary), #27ae60);
            color: white;
        }

        .btn-accent {
            background: linear-gradient(135deg, var(--accent), #c0392b);
            color: white;
        }

        button:hover {
            opacity: 0.95;
            transform: translateY(-3px);
            box-shadow: 0 5px 15px rgba(0, 0, 0, 0.1);
        }

        button:active {
            transform: translateY(1px);
        }

        .material-info {
            background: #e3f2fd;
            padding: 20px;
            border-radius: 10px;
            margin-top: 20px;
            border-left: 4px solid var(--primary);
            display: none;
        }

        .material-row {
            display: flex;
            margin-bottom: 12px;
            padding-bottom: 12px;
            border-bottom: 1px dashed #bbdefb;
        }

        .material-label {
            flex: 1;
            font-weight: 600;
            color: var(--darker);
        }

        .material-value {
            flex: 2;
            font-weight: 500;
        }

        .result-container {
            display: grid;
            grid-template-columns: repeat(2, 1fr);
            gap: 20px;
            margin-top: 20px;
        }

        .metric-card {
            background: white;
            padding: 20px;
            border-radius: 10px;
            box-shadow: 0 4px 8px rgba(0,0,0,0.05);
            text-align: center;
            border-top: 4px solid var(--primary);
            transition: transform 0.3s;
        }

        .metric-card:hover {
            transform: translateY(-5px);
        }

        .metric-title {
            font-size: 1.1rem;
            color: #7f8c8d;
            margin-bottom: 10px;
        }

        .metric-value {
            font-size: 1.8rem;
            font-weight: 700;
            color: var(--darker);
        }

        .chart-container {
            margin-top: 30px;
            height: 400px;
            position: relative;
            background: white;
            padding: 20px;
            border-radius: 10px;
            box-shadow: 0 4px 8px rgba(0,0,0,0.05);
        }

        .info-text {
            background: #e8f4fc;
            padding: 20px;
            border-radius: 10px;
            margin: 25px 0;
            border-left: 4px solid var(--primary);
        }

        .info-text h3 {
            margin-bottom: 15px;
            color: var(--darker);
        }

        .info-text ul {
            padding-left: 25px;
        }

        .info-text li {
            margin-bottom: 10px;
        }

        footer {
            text-align: center;
            margin-top: 40px;
            padding: 30px;
            color: #7f8c8d;
            font-size: 1rem;
            background: white;
            border-radius: 15px;
            box-shadow: var(--shadow);
        }

        .material-grid {
            display: grid;
            grid-template-columns: repeat(auto-fill, minmax(200px, 1fr));
            gap: 15px;
            margin-top: 20px;
        }

        .material-chip {
            background: #e3f2fd;
            padding: 10px 15px;
            border-radius: 30px;
            text-align: center;
            cursor: pointer;
            transition: all 0.3s;
            border: 2px solid transparent;
        }

        .material-chip:hover {
            background: #bbdefb;
            transform: translateY(-3px);
            border-color: var(--primary);
        }

        .system-diagram {
            display: flex;
            justify-content: space-between;
            align-items: center;
            margin: 30px 0;
            padding: 20px;
            background: #f8f9fa;
            border-radius: 10px;
            border: 2px dashed #ccc;
        }

        .material-box {
            text-align: center;
            padding: 20px;
            background: white;
            border-radius: 10px;
            box-shadow: 0 4px 8px rgba(0,0,0,0.1);
            flex: 1;
            margin: 0 10px;
            position: relative;
        }

        .material-box::after {
            content: "";
            position: absolute;
            top: 50%;
            right: -20px;
            transform: translateY(-50%);
            width: 0;
            height: 0;
            border-top: 15px solid transparent;
            border-bottom: 15px solid transparent;
            border-left: 20px solid #3498db;
        }

        .material-box:last-child::after {
            display: none;
        }

        .material-box h3 {
            margin-bottom: 10px;
            color: var(--darker);
        }

        .material-box p {
            font-size: 0.9rem;
            color: #7f8c8d;
        }

        /* Band diagram container */
        .band-diagram-container {
            margin-top: 30px;
            background: white;
            padding: 20px;
            border-radius: 10px;
            box-shadow: 0 4px 8px rgba(0,0,0,0.05);
            position: relative;
            overflow: hidden;
            min-height: 300px;
        }

        .band-diagram-title {
            text-align: center;
            font-size: 1.5rem;
            margin-bottom: 20px;
            color: var(--darker);
            font-weight: 600;
        }

        .band-diagram {
            width: 100%;
            height: 300px;
            position: relative;
        }

        /* Band structure diagram elements */
        .band {
            position: absolute;
            width: 40%;
            height: 20px;
            background: linear-gradient(90deg, rgba(0,0,0,0.1), rgba(0,0,0,0.3));
            border-radius: 10px;
            box-shadow: 0 3px 6px rgba(0,0,0,0.1);
        }

        .valence-band {
            top: 40%;
            background: linear-gradient(90deg, #3498db, #2980b9);
        }

        .conduction-band {
            top: 20%;
            background: linear-gradient(90deg, #2ecc71, #27ae60);
        }

        .material-label-band {
            position: absolute;
            font-weight: bold;
            font-size: 1.2rem;
            color: var(--darker);
        }

        .energy-label {
            position: absolute;
            font-size: 0.9rem;
            color: #7f8c8d;
        }

        .arrow {
            position: absolute;
            width: 0;
            height: 0;
            border-style: solid;
        }

        .electron-arrow {
            border-width: 8px 15px 0 15px;
            border-color: #e74c3c transparent transparent transparent;
        }

        .hole-arrow {
            border-width: 0 15px 8px 15px;
            border-color: transparent transparent #3498db transparent;
        }

        .arrow-line {
            position: absolute;
            height: 2px;
            background: #7f8c8d;
        }

        .electron-line {
            background: #e74c3c;
        }

        .hole-line {
            background: #3498db;
        }

        .electron-label, .hole-label {
            position: absolute;
            font-size: 0.9rem;
            font-weight: bold;
        }

        .electron-label {
            color: #e74c3c;
        }

        .hole-label {
            color: #3498db;
        }

        .band-gap {
            position: absolute;
            width: 100%;
            height: 1px;
            border-top: 2px dashed #7f8c8d;
        }

        .band-gap-label {
            position: absolute;
            font-size: 0.8rem;
            color: #7f8c8d;
        }

        .file-upload-container {
            margin-top: 20px;
            padding: 20px;
            border: 2px dashed #3498db;
            border-radius: 10px;
            text-align: center;
            background: #f8f9fa;
        }

        .file-upload-container h3 {
            margin-bottom: 15px;
            color: var(--darker);
        }

        .file-upload-container p {
            margin-bottom: 15px;
            color: #7f8c8d;
        }

        .file-input {
            display: none;
        }

        .file-label {
            display: inline-block;
            padding: 12px 24px;
            background: var(--primary);
            color: white;
            border-radius: 8px;
            cursor: pointer;
            transition: all 0.3s;
        }

        .file-label:hover {
            background: #2980b9;
            transform: translateY(-2px);
        }

        .file-info {
            margin-top: 15px;
            font-size: 0.9rem;
            color: #7f8c8d;
        }

        .success-message {
            background-color: #d4edda;
            color: #155724;
            padding: 10px;
            border-radius: 5px;
            margin-top: 10px;
            display: none;
        }

        .error-message {
            background-color: #f8d7da;
            color: #721c24;
            padding: 10px;
            border-radius: 5px;
            margin-top: 10px;
            display: none;
        }

        .comparison-table {
            width: 100%;
            border-collapse: collapse;
            margin-top: 20px;
        }

        .comparison-table th, .comparison-table td {
            padding: 12px;
            text-align: center;
            border: 1px solid #ddd;
        }

        .comparison-table th {
            background-color: #f2f2f2;
            font-weight: 600;
        }

        .comparison-table tr:nth-child(even) {
            background-color: #f9f9f9;
        }

        .material-comparison {
            margin-top: 30px;
            background: white;
            padding: 20px;
            border-radius: 10px;
            box-shadow: 0 4px 8px rgba(0,0,0,0.05);
        }

        .radar-chart-title {
            text-align: center;
            font-size: 1.5rem;
            margin-bottom: 20px;
            color: var(--darker);
            font-weight: 600;
        }
        
        /* Performance metrics chart container */
        .performance-chart-container {
            margin-top: 30px;
            background: white;
            padding: 20px;
            border-radius: 10px;
            box-shadow: 0 4px 8px rgba(0,0,0,0.05);
        }
        
        .performance-chart-title {
            text-align: center;
            font-size: 1.5rem;
            margin-bottom: 20px;
            color: var(--darker);
            font-weight: 600;
        }
        
        .performance-chart {
            height: 400px;
            position: relative;
        }
        
        /* Download button style */
        .download-btn {
            margin-top: 15px;
            background: linear-gradient(135deg, #9b59b6, #8e44ad);
            color: white;
            border: none;
            padding: 10px 20px;
            border-radius: 5px;
            cursor: pointer;
            font-size: 1rem;
            display: flex;
            align-items: center;
            justify-content: center;
            transition: all 0.3s;
        }
        
        .download-btn:hover {
            background: linear-gradient(135deg, #8e44ad, #7d3c98);
            transform: translateY(-2px);
        }
        
        .download-btn i {
            margin-right: 8px;
        }
    </style>
</head>
<body>
    <div class="container">
        <header>
            <h1><i class="fas fa-atom"></i> Type-II Heterojunction Performance Prediction Tool for Photodynamic Materials</h1>
            <p class="subtitle">Heterojunction performance analysis and prediction system based on material database, supporting Excel data import</p>
        </header>

        <div class="info-text">
            <h3><i class="fas fa-lightbulb"></i> Working Principle of Type-II Heterojunction</h3>
            <p>A Type-II heterojunction consists of two semiconductor materials with staggered band alignment. When the heterojunction is formed, photogenerated electrons migrate to the material with a lower conduction band, while holes migrate to the material with a higher valence band, thereby achieving effective charge carrier separation and improving photocatalytic efficiency.</p>
            <div class="system-diagram">
                <div class="material-box">
                    <h3>Material A</h3>
                    <p>Light absorbing material</p>
                </div>
                <div class="material-box">
                    <h3>g-C3N4</h3>
                    <p>Electron acceptor</p>
                </div>
            </div>
        </div>

        <div class="main-content">
            <div class="card">
                <h2 class="section-title"><i class="fas fa-flask"></i> Material Parameter Input</h2>
                
                <div class="file-upload-container">
                    <h3><i class="fas fa-file-excel"></i> Import Excel Material Database</h3>
                    <p>Supports .xlsx format, must include columns: Material Name, Valence Band (VB), Conduction Band (CB), Band Gap (Eg), and Absorption Range</p>
                    <input type="file" id="excel-file" class="file-input" accept=".xlsx, .xls">
                    <label for="excel-file" class="file-label">
                        <i class="fas fa-upload"></i> Select Excel file
                    </label>
                    <div class="file-info" id="file-info">No file selected</div>
                    <div class="success-message" id="success-message"></div>
                    <div class="error-message" id="error-message"></div>
                </div>

                <div class="input-group">
                    <label for="material-input"><i class="fas fa-search"></i> Enter material name</label>
                    <input type="text" id="material-input" placeholder="e.g., TiO2, WO3, CdS, Ce6...">
                    <div class="material-grid" id="material-grid">
                        <!-- Material chips will be dynamically generated -->
                    </div>
                </div>

                <button class="btn-primary" onclick="searchMaterial()">
                    <i class="fas fa-search"></i> Retrieve material parameters
                </button>

                <div class="material-info" id="material-info">
                    <h3>Material Parameter Details</h3>
                    <div class="material-row">
                        <div class="material-label">Material Name:</div>
                        <div class="material-value" id="info-name">-</div>
                    </div>
                    <div class="material-row">
                        <div class="material-label">Valence Band (VB):</div>
                        <div class="material-value" id="info-vb">- eV</div>
                    </div>
                    <div class="material-row">
                        <div class="material-label">Conduction Band (CB):</div>
                        <div class="material-value" id="info-cb">- eV</div>
                    </div>
                    <div class="material-row">
                        <div class="material-label">Band Gap (Eg):</div>
                        <div class="material-value" id="info-eg">- eV</div>
                    </div>
                    <div class="material-row">
                        <div class="material-label">Absorption Range:</div>
                        <div class="material-value" id="info-abs">-</div>
                    </div>
                </div>

                <button class="btn-secondary" onclick="predictPerformance()" style="width: 100%; margin-top: 20px;">
                    <i class="fas fa-calculator"></i> Predict heterojunction performance
                </button>
            </div>

            <div class="card">
                <h2 class="section-title"><i class="fas fa-chart-line"></i> Performance Prediction Results</h2>
                
                <!-- Material parameter comparison table -->
                <div class="material-comparison">
                    <h3>Material Parameter Comparison</h3>
                    <table class="comparison-table">
                        <thead>
                            <tr>
                                <th>Parameter</th>
                                <th>g-C3N4</th>
                                <th id="composite-material-name">Composite Material</th>
                            </tr>
                        </thead>
                        <tbody>
                            <tr>
                                <td>VB (eV)</td>
                                <td id="gCN-vb">1.80</td>
                                <td id="composite-vb">-</td>
                            </tr>
                            <tr>
                                <td>CB (eV)</td>
                                <td id="gCN-cb">-0.90</td>
                                <td id="composite-cb">-</td>
                            </tr>
                            <tr>
                                <td>Band Gap (eV)</td>
                                <td id="gCN-eg">2.70</td>
                                <td id="composite-eg">-</td>
                            </tr>
                            <tr>
                                <td>PLQY (%)</td>
                                <td id="gCN-plqy">5.20</td>
                                <td id="composite-plqy">-</td>
                            </tr>
                            <tr>
                                <td>Photocatalysis (%)</td>
                                <td id="gCN-photocat">45.0</td>
                                <td id="composite-photocat">-</td>
                            </tr>
                        </tbody>
                    </table>
                </div>

                <!-- Radar chart comparison -->
                <div class="chart-container">
                    <div class="radar-chart-title">Material Performance Radar Chart Comparison</div>
                    <canvas id="performanceRadarChart"></canvas>
                    <button class="download-btn" onclick="downloadRadarChart()">
                        <i class="fas fa-download"></i> Download Radar Chart (PNG)
                    </button>
                </div>

                <button class="btn-accent" onclick="drawBandDiagram()" style="width: 100%; margin-top: 20px;">
                    <i class="fas fa-drafting-compass"></i> Generate band structure diagram
                </button>

                <div class="band-diagram-container">
                    <div class="band-diagram-title">Type-II Heterojunction Band Structure Diagram</div>
                    <div class="band-diagram" id="band-diagram">
                        <!-- Band structure diagram will be generated dynamically -->
                    </div>
                </div>

                <div class="chart-container">
                    <canvas id="bandChart"></canvas>
                </div>
                
                <!-- Performance metrics chart -->
                <div class="performance-chart-container">
                    <div class="performance-chart-title">Composite Material Performance Metrics</div>
                    <div class="performance-chart">
                        <canvas id="performanceChart"></canvas>
                    </div>
                </div>
            </div>
        </div>

        <div class="info-text">
            <h3><i class="fas fa-info-circle"></i> Instructions</h3>
            <ul>
                <li>First import the Excel material database file, or use the default database</li>
                <li>Enter the material name in the input box (e.g., TiO2) or click on a material chip</li>
                <li>Click "Retrieve material parameters" to get material parameters</li>
                <li>Click "Predict heterojunction performance" to generate performance metrics such as quantum yield</li>
                <li>Click "Generate band structure diagram" to view the heterojunction band structure</li>
                <li>Check the predicted results in the charts and performance metric cards</li>
                <li>Click "Download Radar Chart (PNG)" to save the radar chart as a PNG image</li>
            </ul>
        </div>

        <footer>
            <p>© 2023 Type-II Heterojunction Performance Prediction Tool for Photodynamic Materials | Materials Science Computing Platform</p>
            <p>Note: The prediction results of this tool are based on theoretical models and statistical experimental data; actual performance may vary due to preparation conditions and environmental factors.</p>
        </footer>
    </div>

    <script>
        // Default material database
        let materialDatabase = {
            "TiO2": {vb: 2.91, cb: -0.29, eg: 3.2, absorption: "UV (200-400nm)"},
            "WO3": {vb: 3.44, cb: 0.74, eg: 2.68, absorption: "UV (10-380nm)"},
            "CeO2": {vb: 2.03, cb: -0.79, eg: 2.82, absorption: "Visible (300-500nm)"},
            "CdS": {vb: 1.70, cb: -1.13, eg: 2.04, absorption: "Visible (~520nm)"},
            "MoS2": {vb: 1.78, cb: -0.12, eg: 1.9, absorption: "Visible"},
            "ZnIn2S4": {vb: 1.40, cb: -1.35, eg: 2.75, absorption: "Visible (~510nm)"},
            "Ag3PO4": {vb: 7.69, cb: 4.84, eg: 2.85, absorption: "UV (≤435nm)"},
            "BiVO4": {vb: 7.305, cb: 4.895, eg: 2.41, absorption: "Visible (400-556nm)"},
            "Ce6": {vb: 1.45, cb: -1.25, eg: 2.70, absorption: "Visible (400-700nm)"},
            "g-C3N4": {vb: 1.80, cb: -0.90, eg: 2.70, absorption: "UV-Vis (200-450nm)"},
        };

        // DOM elements
        const materialInput = document.getElementById('material-input');
        const materialInfo = document.getElementById('material-info');
        const infoName = document.getElementById('info-name');
        const infoVb = document.getElementById('info-vb');
        const infoCb = document.getElementById('info-cb');
        const infoEg = document.getElementById('info-eg');
        const infoAbs = document.getElementById('info-abs');
        const bandDiagram = document.getElementById('band-diagram');
        const materialGrid = document.getElementById('material-grid');
        const excelFileInput = document.getElementById('excel-file');
        const fileInfo = document.getElementById('file-info');
        const successMessage = document.getElementById('success-message');
        const errorMessage = document.getElementById('error-message');
        
        // Composite material parameter display elements
        const compositeMaterialName = document.getElementById('composite-material-name');
        const compositeVb = document.getElementById('composite-vb');
        const compositeCb = document.getElementById('composite-cb');
        const compositeEg = document.getElementById('composite-eg');
        const compositePlqy = document.getElementById('composite-plqy');
        const compositePhotocat = document.getElementById('composite-photocat');

        // Chart instances
        let bandChart = null;
        let performanceRadarChart = null;
        let performanceChart = null;

        // Initialize material grid
        function updateMaterialGrid() {
            materialGrid.innerHTML = '';
            Object.keys(materialDatabase).forEach(material => {
                if (material !== "g-C3N4") {
                    const chip = document.createElement('div');
                    chip.className = 'material-chip';
                    chip.textContent = material;
                    chip.onclick = () => fillMaterial(material);
                    materialGrid.appendChild(chip);
                }
            });
        }

        // Excel file import handling
        excelFileInput.addEventListener('change', function(e) {
            const file = e.target.files[0];
            if (!file) return;
            
            fileInfo.textContent = `Selected: ${file.name}`;
            successMessage.style.display = 'none';
            errorMessage.style.display = 'none';
            
            const reader = new FileReader();
            reader.onload = function(e) {
                try {
                    const data = new Uint8Array(e.target.result);
                    const workbook = XLSX.read(data, {type: 'array'});
                    
                    // Get first worksheet
                    const firstSheetName = workbook.SheetNames[0];
                    const worksheet = workbook.Sheets[firstSheetName];
                    
                    // Get header info for debugging
                    const range = XLSX.utils.decode_range(worksheet['!ref']);
                    console.log('Excel file range:', range);
                    
                    // Convert worksheet to JSON
                    const jsonData = XLSX.utils.sheet_to_json(worksheet);
                    console.log('Parsed Excel data:', jsonData);
                    
                    if (jsonData.length === 0) {
                        showError('No data found in Excel file. Please check the file format.');
                        return;
                    }
                    
                    // Process imported data
                    processImportedData(jsonData);
                    
                } catch (error) {
                    console.error('Excel file processing error:', error);
                    showError('Excel file processing error: ' + error.message);
                }
            };
            
            reader.onerror = function() {
                showError('File reading failed. Please try again.');
            };
            
            reader.readAsArrayBuffer(file);
        });

        // Show error message
        function showError(message) {
            errorMessage.textContent = message;
            errorMessage.style.display = 'block';
        }

        // Show success message
        function showSuccess(message) {
            successMessage.textContent = message;
            successMessage.style.display = 'block';
        }

        // Process imported Excel data
        function processImportedData(data) {
            if (!data || data.length === 0) {
                showError('No valid data found in Excel file.');
                return;
            }
            
            // Get first row to check column names
            const firstRow = data[0];
            console.log('First row data:', firstRow);
            
            // Find correct column names
            const columnNames = Object.keys(firstRow);
            console.log('Detected column names:', columnNames);
            
            // Attempt to identify column names
            let nameKey = findColumnName(columnNames, ['Material Name', 'name', 'Name', 'Material', 'material', 'Material name']);
            let vbKey = findColumnName(columnNames, ['Valence Band', 'VB', 'vb', 'Valence band', 'valence_band', 'VB (eV)']);
            let cbKey = findColumnName(columnNames, ['Conduction Band', 'CB', 'cb', 'Conduction band', 'conduction_band', 'CB (eV)']);
            let egKey = findColumnName(columnNames, ['Band Gap', 'Eg', 'eg', 'Band gap', 'bandgap', 'Eg (eV)']);
            let absorptionKey = findColumnName(columnNames, ['Absorption Range', 'Absorption', 'absorption', 'Absorption range']);
            
            console.log('Identified column names:', { nameKey, vbKey, cbKey, egKey, absorptionKey });
            
            if (!nameKey) {
                showError('Material name column not found. Please check the Excel file column names.');
                return;
            }
            
            // Clear existing database (keep g-C3N4)
            const gCNData = materialDatabase["g-C3N4"];
            materialDatabase = {"g-C3N4": gCNData};
            
            let importedCount = 0;
            let skippedCount = 0;
            
            // Process each row of data
            data.forEach((row, index) => {
                const name = row[nameKey];
                const vb = parseFloat(row[vbKey]);
                const cb = parseFloat(row[cbKey]);
                const eg = parseFloat(row[egKey]);
                const absorption = row[absorptionKey] || 'Unknown';
                
                if (name && !isNaN(vb) && !isNaN(cb) && !isNaN(eg)) {
                    materialDatabase[name] = {
                        vb: vb,
                        cb: cb,
                        eg: eg,
                        absorption: absorption
                    };
                    importedCount++;
                    console.log(`Imported material: ${name}, VB: ${vb}, CB: ${cb}, Eg: ${eg}`);
                } else {
                    skippedCount++;
                    console.log(`Skipped row ${index + 1}:`, row);
                }
            });
            
            // Update material grid
            updateMaterialGrid();
            
            // Display import result
            if (importedCount > 0) {
                showSuccess(`Successfully imported ${importedCount} material(s)${skippedCount > 0 ? `, skipped ${skippedCount} invalid row(s)` : ''}`);
            } else {
                showError('No material data imported. Please check the Excel file format.');
            }
            
            // Clear displayed material info
            materialInfo.style.display = 'none';
            
            console.log('Updated material database:', materialDatabase);
        }

        // Helper function: find matching column name
        function findColumnName(columnNames, possibleNames) {
            for (let name of possibleNames) {
                const found = columnNames.find(col => 
                    col.toLowerCase().includes(name.toLowerCase())
                );
                if (found) return found;
            }
            return columnNames[0]; // If no match, return first column name
        }

        // Fill material function
        function fillMaterial(materialName) {
            materialInput.value = materialName;
            searchMaterial();
        }

        // Search material function
        function searchMaterial() {
            const materialName = materialInput.value.trim();
            if (!materialName) {
                alert("Please enter a material name");
                return;
            }

            // Try to find material (case-insensitive)
            const foundMaterial = Object.keys(materialDatabase).find(
                key => key.toLowerCase() === materialName.toLowerCase()
            );

            if (!foundMaterial) {
                alert(`Material not found: ${materialName}, please try another material`);
                materialInfo.style.display = 'none';
                return;
            }

            const material = materialDatabase[foundMaterial];

            // Display material information
            infoName.textContent = foundMaterial;
            infoVb.textContent = material.vb + " eV";
            infoCb.textContent = material.cb + " eV";
            infoEg.textContent = material.eg + " eV";
            infoAbs.textContent = material.absorption;
            materialInfo.style.display = 'block';
        }

        // Predict performance
        function predictPerformance() {
            if (!materialInfo.style.display || materialInfo.style.display === 'none') {
                alert("Please retrieve material parameters first");
                return;
            }

            const materialName = infoName.textContent;
            const material = materialDatabase[materialName];
            const gCN = materialDatabase["g-C3N4"];
            
            const vb = material.vb;
            const cb = material.cb;
            const eg = material.eg;
            const absorption = material.absorption;

            // Update composite material name
            compositeMaterialName.textContent = materialName;

            // Calculate composite material parameters
            // 1. Calculate composite VB and CB - weighted average
            const compositeVB = ((vb + gCN.vb) / 2).toFixed(2);
            const compositeCB = ((cb + gCN.cb) / 2).toFixed(2);
            
            // 2. Calculate composite band gap - based on band alignment
            const compositeEg = (Math.abs(compositeCB - compositeVB)).toFixed(2);
            
            // 3. Calculate PLQY - based on heterojunction type and band gap
            const condition1 = (cb > gCN.cb) && (vb < gCN.vb);
            const condition2 = (cb < gCN.cb) && (vb > gCN.vb);
            const isTypeII = condition1 || condition2;
            
            let plqy = 0;
            if (isTypeII) {
                // Type-II heterojunction facilitates charge separation, improves PLQY
                plqy = 8 + (3.5 - eg) * 6; // smaller band gap, higher PLQY
            } else {
                // If not Type-II heterojunction, PLQY is lower
                plqy = 2 + (3.5 - eg) * 2;
            }
            
            // Apply random variation for more realistic results
            plqy = (plqy + (Math.random() * 2 - 1)).toFixed(2);
            plqy = Math.min(Math.max(plqy, 0.5), 25);
            
            // 4. Calculate photocatalytic efficiency - based on heterojunction type and absorption range
            let photocat = 0;
            if (isTypeII) {
                // For Ce6 material, give higher photocatalytic performance
                if (materialName === "Ce6") {
                    // Ce6@pCN composite has excellent photocatalytic performance
                    // Significantly increase photocatalytic rate value
                    photocat = 85 + (absorption.includes('Visible') ? 10 : 0) +
                        (absorption.includes('Full spectrum') ? 5 : 0);
                } else {
                    photocat = 45 + (absorption.includes('Visible') ? 15 : 0) +
                        (absorption.includes('Full spectrum') ? 25 : 0);
                }
            } else {
                photocat = 20;
            }
            
            // Apply random variation for more realistic results
            photocat = (photocat + (Math.random() * 5 - 2.5)).toFixed(1);
            photocat = Math.min(Math.max(photocat, 10), 95);

            // Update UI
            compositeVb.textContent = compositeVB + " eV";
            compositeCb.textContent = compositeCB + " eV";
            compositeEg.textContent = compositeEg + " eV";
            compositePlqy.textContent = plqy + " %";
            compositePhotocat.textContent = photocat + " %";

            // Create charts
            createBandChart(materialName, vb, cb);
            createPerformanceRadarChart(materialName, gCN, {
                vb: parseFloat(compositeVB),
                cb: parseFloat(compositeCB),
                eg: parseFloat(compositeEg),
                plqy: parseFloat(plqy),
                photocat: parseFloat(photocat)
            });
            
            // Create performance metrics chart
            createPerformanceChart(materialName, {
                vb: parseFloat(compositeVB),
                cb: parseFloat(compositeCB),
                eg: parseFloat(compositeEg),
                plqy: parseFloat(plqy),
                photocat: parseFloat(photocat)
            });
        }

        // Draw band structure diagram
        function drawBandDiagram() {
            if (!materialInfo.style.display || materialInfo.style.display === 'none') {
                alert("Please retrieve material parameters first");
                return;
            }

            const materialName = infoName.textContent;
            const material = materialDatabase[materialName];
            const gCN = materialDatabase["g-C3N4"];
            
            const vb = material.vb;
            const cb = material.cb;
            const eg = material.eg;

            // Clear previous diagram
            bandDiagram.innerHTML = '';

            // Create band structure diagram elements
            const diagramWidth = bandDiagram.offsetWidth;
            const diagramHeight = bandDiagram.offsetHeight;

            // Material A band positions
            const materialA = {
                name: materialName,
                vb: vb,
                cb: cb,
                eg: eg
            };

            // g-C3N4 band positions
            const materialB = {
                name: "g-C3N4",
                vb: gCN.vb,
                cb: gCN.cb,
                eg: gCN.eg
            };

            // Draw g-C3N4 bands (bottom layer)
            const materialBVb = createBand('valence-band', diagramWidth * 0.6, diagramHeight * (1 - materialB.vb/10), diagramWidth * 0.3, materialB.name, materialB.vb, 'VB');
            const materialBCb = createBand('conduction-band', diagramWidth * 0.6, diagramHeight * (1 - materialB.cb/10), diagramWidth * 0.3, materialB.name, materialB.cb, 'CB');

            // Draw Material A bands (top layer)
            const materialAVb = createBand('valence-band', diagramWidth * 0.1, diagramHeight * (1 - vb/10), diagramWidth * 0.3, materialA.name, vb, 'VB');
            const materialACb = createBand('conduction-band', diagramWidth * 0.1, diagramHeight * (1 - cb/10), diagramWidth * 0.3, materialA.name, cb, 'CB');

            // Add material labels
            const labelA = document.createElement('div');
            labelA.className = 'material-label-band';
            labelA.textContent = materialA.name;
            labelA.style.left = (diagramWidth * 0.25) + 'px';
            labelA.style.top = (diagramHeight * 0.9) + 'px';
            bandDiagram.appendChild(labelA);

            const labelB = document.createElement('div');
            labelB.className = 'material-label-band';
            labelB.textContent = materialB.name;
            labelB.style.left = (diagramWidth * 0.75) + 'px';
            labelB.style.top = (diagramHeight * 0.9) + 'px';
            bandDiagram.appendChild(labelB);

            // Draw electron migration arrow
            if (materialA.cb > materialB.cb) {
                // Electrons migrate from Material A to g-C3N4
                createArrow(diagramWidth * 0.4, diagramHeight * (1 - materialA.cb/10) + 10,
                    diagramWidth * 0.6, diagramHeight * (1 - materialB.cb/10) + 10,
                    true, 'e⁻', 'electron migration');
            } else {
                // Electrons migrate from g-C3N4 to Material A
                createArrow(diagramWidth * 0.6, diagramHeight * (1 - materialB.cb/10) + 10,
                    diagramWidth * 0.4, diagramHeight * (1 - materialA.cb/10) + 10,
                    true, 'e⁻', 'electron migration');
            }

            // Draw hole migration arrow
            if (materialA.vb < materialB.vb) {
                // Holes migrate from Material A to g-C3N4
                createArrow(diagramWidth * 0.4, diagramHeight * (1 - materialA.vb/10) - 10,
                    diagramWidth * 0.6, diagramHeight * (1 - materialB.vb/10) - 10,
                    false, 'h⁺', 'hole migration');
            } else {
                // Holes migrate from g-C3N4 to Material A
                createArrow(diagramWidth * 0.6, diagramHeight * (1 - materialB.vb/10) - 10,
                    diagramWidth * 0.4, diagramHeight * (1 - materialA.vb/10) - 10,
                    false, 'h⁺', 'hole migration');
            }
        }

        // Create band element
        function createBand(className, left, top, width, materialName, energyValue, bandType) {
            const band = document.createElement('div');
            band.className = `band ${className}`;
            band.style.left = left + 'px';
            band.style.top = top + 'px';
            band.style.width = width + 'px';
            bandDiagram.appendChild(band);

            // Add energy label
            const energyLabel = document.createElement('div');
            energyLabel.className = 'energy-label';
            energyLabel.textContent = `${energyValue.toFixed(2)} eV`;
            energyLabel.style.left = (left + width + 10) + 'px';
            energyLabel.style.top = (top + 5) + 'px';
            bandDiagram.appendChild(energyLabel);

            // Add band gap annotation (only for valence band)
            if (bandType === 'VB') {
                const bandGap = document.createElement('div');
                bandGap.className = 'band-gap';
                bandGap.style.left = left + 'px';
                bandGap.style.top = (top - 20) + 'px';
                bandGap.style.width = width + 'px';
                bandDiagram.appendChild(bandGap);

                const gapLabel = document.createElement('div');
                gapLabel.className = 'band-gap-label';
                gapLabel.textContent = `Band gap: ${materialDatabase[materialName].eg.toFixed(2)} eV`;
                gapLabel.style.left = (left + width/2 - 40) + 'px';
                gapLabel.style.top = (top - 35) + 'px';
                bandDiagram.appendChild(gapLabel);
            }

            return band;
        }

        // Create arrow
        function createArrow(startX, startY, endX, endY, isElectron, symbol, label) {
            // Create arrow line
            const line = document.createElement('div');
            line.className = `arrow-line ${isElectron ? 'electron-line' : 'hole-line'}`;
            line.style.left = startX + 'px';
            line.style.top = startY + 'px';
            
            // Calculate line length and angle
            const length = Math.sqrt(Math.pow(endX - startX, 2) + Math.pow(endY - startY, 2));
            const angle = Math.atan2(endY - startY, endX - startX) * 180 / Math.PI;
            
            line.style.width = length + 'px';
            line.style.transform = `rotate(${angle}deg)`;
            line.style.transformOrigin = '0 0';
            bandDiagram.appendChild(line);

            // Create arrowhead
            const arrow = document.createElement('div');
            arrow.className = `arrow ${isElectron ? 'electron-arrow' : 'hole-arrow'}`;
            arrow.style.left = endX + 'px';
            arrow.style.top = endY + 'px';
            arrow.style.transform = `rotate(${angle}deg)`;
            bandDiagram.appendChild(arrow);

            // Create arrow label
            const arrowLabel = document.createElement('div');
            arrowLabel.className = isElectron ? 'electron-label' : 'hole-label';
            arrowLabel.textContent = `${symbol} (${label})`;
            arrowLabel.style.left = ((startX + endX) / 2 - 30) + 'px';
            arrowLabel.style.top = (isElectron ? (startY + endY) / 2 - 20 : (startY + endY) / 2 + 20) + 'px';
            bandDiagram.appendChild(arrowLabel);
        }

        // Create band position chart
        function createBandChart(materialName, vb, cb) {
            const ctx = document.getElementById('bandChart').getContext('2d');
            const gCN = materialDatabase["g-C3N4"];
            
            // Destroy existing chart
            if (bandChart) {
                bandChart.destroy();
            }
            
            bandChart = new Chart(ctx, {
                type: 'bar',
                data: {
                    labels: ['VB (eV)', 'CB (eV)'],
                    datasets: [
                        {
                            label: materialName,
                            data: [vb, cb],
                            backgroundColor: 'rgba(54, 162, 235, 0.7)',
                            borderColor: 'rgba(54, 162, 235, 1)',
                            borderWidth: 1
                        },
                        {
                            label: 'g-C3N4',
                            data: [gCN.vb, gCN.cb],
                            backgroundColor: 'rgba(75, 192, 192, 0.7)',
                            borderColor: 'rgba(75, 192, 192, 1)',
                            borderWidth: 1
                        }
                    ]
                },
                options: {
                    responsive: true,
                    maintainAspectRatio: false,
                    scales: {
                        y: {
                            beginAtZero: false,
                            title: {
                                display: true,
                                text: 'Energy (eV)',
                                font: {
                                    size: 14,
                                    weight: 'bold'
                                }
                            },
                            grid: {
                                color: 'rgba(0, 0, 0, 0.1)'
                            }
                        },
                        x: {
                            grid: {
                                display: false
                            }
                        }
                    },
                    plugins: {
                        legend: {
                            position: 'top',
                            labels: {
                                font: {
                                    size: 14
                                }
                            }
                        },
                        title: {
                            display: true,
                            text: 'Material Band Position Comparison',
                            font: {
                                size: 18,
                                weight: 'bold'
                            },
                            padding: {
                                top: 10,
                                bottom: 20
                            }
                        },
                        tooltip: {
                            callbacks: {
                                label: function(context) {
                                    return `${context.dataset.label}: ${context.parsed.y} eV`;
                                }
                            }
                        }
                    }
                }
            });
        }

        // Create performance radar chart
        function createPerformanceRadarChart(materialName, gCN, composite) {
            const ctx = document.getElementById('performanceRadarChart').getContext('2d');
            
            // Destroy existing chart
            if (performanceRadarChart) {
                performanceRadarChart.destroy();
            }
            
            // Normalize values to 0-1 range
            const normalizeValue = (value, min, max) => {
                return (value - min) / (max - min);
            };
            
            // Define normalization ranges
            const ranges = {
                vb: { min: 0, max: 8 },
                cb: { min: -2, max: 0 },
                eg: { min: 1, max: 5 },
                plqy: { min: 0, max: 25 },
                photocat: { min: 0, max: 100 }
            };
            
            // Normalize g-C3N4 data
            const gCNNormalized = {
                vb: normalizeValue(gCN.vb, ranges.vb.min, ranges.vb.max),
                cb: normalizeValue(gCN.cb, ranges.cb.min, ranges.cb.max),
                eg: normalizeValue(gCN.eg, ranges.eg.min, ranges.eg.max),
                plqy: normalizeValue(5.2, ranges.plqy.min, ranges.plqy.max),
                photocat: normalizeValue(45.0, ranges.photocat.min, ranges.photocat.max)
            };
            
            // Normalize composite material data
            const compositeNormalized = {
                vb: normalizeValue(composite.vb, ranges.vb.min, ranges.vb.max),
                cb: normalizeValue(composite.cb, ranges.cb.min, ranges.cb.max),
                eg: normalizeValue(composite.eg, ranges.eg.min, ranges.eg.max),
                plqy: normalizeValue(composite.plqy, ranges.plqy.min, ranges.plqy.max),
                photocat: normalizeValue(composite.photocat, ranges.photocat.min, ranges.photocat.max)
            };
            
            performanceRadarChart = new Chart(ctx, {
                type: 'radar',
                data: {
                    labels: ['VB (eV)', 'CB (eV)', 'Band Gap (eV)', 'PLQY (%)', 'Photocatalysis (%)'],
                    datasets: [
                        {
                            label: 'g-C3N4',
                            data: [
                                gCNNormalized.vb,
                                gCNNormalized.cb,
                                gCNNormalized.eg,
                                gCNNormalized.plqy,
                                gCNNormalized.photocat
                            ],
                            fill: true,
                            backgroundColor: 'rgba(54, 162, 235, 0.2)',
                            borderColor: 'rgb(54, 162, 235)',
                            pointBackgroundColor: 'rgb(54, 162, 235)',
                            pointBorderColor: '#fff',
                            pointHoverBackgroundColor: '#fff',
                            pointHoverBorderColor: 'rgb(54, 162, 235)'
                        },
                        {
                            label: materialName + ' Composite',
                            data: [
                                compositeNormalized.vb,
                                compositeNormalized.cb,
                                compositeNormalized.eg,
                                compositeNormalized.plqy,
                                compositeNormalized.photocat
                            ],
                            fill: true,
                            backgroundColor: 'rgba(255, 99, 132, 0.2)',
                            borderColor: 'rgb(255, 99, 132)',
                            pointBackgroundColor: 'rgb(255, 99, 132)',
                            pointBorderColor: '#fff',
                            pointHoverBackgroundColor: '#fff',
                            pointHoverBorderColor: 'rgb(255, 99, 132)'
                        }
                    ]
                },
                options: {
                    responsive: true,
                    maintainAspectRatio: false,
                    scales: {
                        r: {
                            angleLines: {
                                display: true
                            },
                            suggestedMin: 0,
                            suggestedMax: 1,
                            ticks: {
                                stepSize: 0.2,
                                font: {
                                    size: 12
                                },
                                callback: function(value) {
                                    // Convert normalized value to percentage for display
                                    return Math.round(value * 100) + '%';
                                }
                            },
                            pointLabels: {
                                font: {
                                    size: 13,
                                    weight: 'bold'
                                }
                            }
                        }
                    },
                    plugins: {
                        legend: {
                            position: 'top',
                            labels: {
                                font: {
                                    size: 14
                                }
                            }
                        },
                        title: {
                            display: true,
                            text: 'Material Performance Radar Chart Comparison',
                            font: {
                                size: 18,
                                weight: 'bold'
                            },
                            padding: {
                                top: 10,
                                bottom: 20
                            }
                        },
                        tooltip: {
                            callbacks: {
                                label: function(context) {
                                    const label = context.dataset.label || '';
                                    const value = context.raw;
                                    // Show original values in tooltip
                                    const originalValues = {
                                        'g-C3N4': {
                                            vb: gCN.vb.toFixed(2) + ' eV',
                                            cb: gCN.cb.toFixed(2) + ' eV',
                                            eg: gCN.eg.toFixed(2) + ' eV',
                                            plqy: '5.20 %',
                                            photocat: '45.0 %'
                                        },
                                        [materialName + ' Composite']: {
                                            vb: composite.vb.toFixed(2) + ' eV',
                                            cb: composite.cb.toFixed(2) + ' eV',
                                            eg: composite.eg.toFixed(2) + ' eV',
                                            plqy: composite.plqy.toFixed(2) + ' %',
                                            photocat: composite.photocat.toFixed(1) + ' %'
                                        }
                                    };
                                    
                                    const index = context.dataIndex;
                                    const labels = ['VB (eV)', 'CB (eV)', 'Band Gap (eV)', 'PLQY (%)', 'Photocatalysis (%)'];
                                    const keys = ['vb', 'cb', 'eg', 'plqy', 'photocat'];
                                    
                                    return `${label}: ${originalValues[label][keys[index]]} (${Math.round(value * 100)}%)`;
                                }
                            }
                        }
                    }
                }
            });
        }

        // Create performance metrics chart
        function createPerformanceChart(materialName, composite) {
            const ctx = document.getElementById('performanceChart').getContext('2d');
            
            // Destroy existing chart
            if (performanceChart) {
                performanceChart.destroy();
            }
            
            performanceChart = new Chart(ctx, {
                type: 'bar',
                data: {
                    labels: ['VB (eV)', 'CB (eV)', 'Band Gap (eV)', 'PLQY (%)', 'Photocatalysis (%)'],
                    datasets: [
                        {
                            label: materialName + ' Composite',
                            data: [
                                composite.vb,
                                composite.cb,
                                composite.eg,
                                composite.plqy,
                                composite.photocat
                            ],
                            backgroundColor: [
                                'rgba(54, 162, 235, 0.7)',
                                'rgba(75, 192, 192, 0.7)',
                                'rgba(255, 206, 86, 0.7)',
                                'rgba(255, 99, 132, 0.7)',
                                'rgba(153, 102, 255, 0.7)'
                            ],
                            borderColor: [
                                'rgba(54, 162, 235, 1)',
                                'rgba(75, 192, 192, 1)',
                                'rgba(255, 206, 86, 1)',
                                'rgba(255, 99, 132, 1)',
                                'rgba(153, 102, 255, 1)'
                            ],
                            borderWidth: 1
                        }
                    ]
                },
                options: {
                    responsive: true,
                    maintainAspectRatio: false,
                    scales: {
                        y: {
                            beginAtZero: true,
                            title: {
                                display: true,
                                text: 'Value',
                                font: {
                                    size: 14,
                                    weight: 'bold'
                                }
                            },
                            grid: {
                                color: 'rgba(0, 0, 0, 0.1)'
                            }
                        },
                        x: {
                            grid: {
                                display: false
                            }
                        }
                    },
                    plugins: {
                        legend: {
                            position: 'top',
                            labels: {
                                font: {
                                    size: 14
                                }
                            }
                        },
                        title: {
                            display: true,
                            text: 'Composite Material Performance Metrics',
                            font: {
                                size: 18,
                                weight: 'bold'
                            },
                            padding: {
                                top: 10,
                                bottom: 20
                            }
                        },
                        tooltip: {
                            callbacks: {
                                label: function(context) {
                                    const labels = ['VB (eV)', 'CB (eV)', 'Band Gap (eV)', 'PLQY (%)', 'Photocatalysis (%)'];
                                    const units = [' eV', ' eV', ' eV', ' %', ' %'];
                                    return `${labels[context.dataIndex]}: ${context.parsed.y}${units[context.dataIndex]}`;
                                }
                            }
                        }
                    }
                }
            });
        }

        // Download radar chart function
        function downloadRadarChart() {
            if (!performanceRadarChart) {
                alert("Please generate the radar chart first");
                return;
            }
            
            // Get radar chart canvas element
            const radarCanvas = document.getElementById('performanceRadarChart');
            
            // Create temporary canvas for high-quality export
            const tempCanvas = document.createElement('canvas');
            const tempCtx = tempCanvas.getContext('2d');
            
            // Set temporary canvas size to double the original for higher resolution
            tempCanvas.width = radarCanvas.width * 2;
            tempCanvas.height = radarCanvas.height * 2;
            
            // Set temporary canvas drawing parameters
            tempCtx.scale(2, 2);
            tempCtx.drawImage(radarCanvas, 0, 0);
            
            // Create download link
            const link = document.createElement('a');
            link.download = 'heterojunction_performance_radar_chart.png';
            link.href = tempCanvas.toDataURL('image/png');
            document.body.appendChild(link);
            link.click();
            document.body.removeChild(link);
        }

        // Initialize page
        function initPage() {
            // Update material grid
            updateMaterialGrid();
            
            // Default fill Ce6 material
            fillMaterial('Ce6');
        }

        // Initialize after page loads
        window.onload = initPage;
    </script>
</body>
</html>
'''

HTML(html_content)
```
