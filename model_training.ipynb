```python
from IPython.display import HTML

html_content = '''
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Heterojunction Composite Iterative Learning System</title>
    <script src="https://cdnjs.cloudflare.com/ajax/libs/xlsx/0.18.5/xlsx.full.min.js"></script>
    <script src="https://cdn.plot.ly/plotly-latest.min.js"></script>
    <style>
        * {
            margin: 0;
            padding: 0;
            box-sizing: border-box;
            font-family: 'Segoe UI', Tahoma, Geneva, Verdana, sans-serif;
        }
        
        body {
            background: linear-gradient(135deg, #1a2a6c, #b21f1f, #fdbb2d);
            color: #333;
            min-height: 100vh;
            padding: 20px;
        }
        
        .container {
            max-width: 1400px;
            margin: 0 auto;
            background: rgba(255, 255, 255, 0.95);
            border-radius: 15px;
            box-shadow: 0 20px 40px rgba(0,0,0,0.2);
            overflow: hidden;
        }
        
        .header {
            background: linear-gradient(135deg, #2c3e50, #3498db);
            color: white;
            padding: 30px;
            text-align: center;
        }
        
        .header h1 {
            font-size: 2.5em;
            margin-bottom: 10px;
            text-shadow: 2px 2px 4px rgba(0,0,0,0.3);
        }
        
        .header p {
            font-size: 1.2em;
            opacity: 0.9;
        }
        
        .content {
            padding: 30px;
        }
        
        .input-panel {
            background: #f8f9fa;
            padding: 25px;
            border-radius: 10px;
            margin-bottom: 30px;
            border-left: 5px solid #3498db;
        }
        
        .input-grid {
            display: grid;
            grid-template-columns: repeat(auto-fit, minmax(200px, 1fr));
            gap: 20px;
            margin-bottom: 20px;
        }
        
        .input-group {
            display: flex;
            flex-direction: column;
        }
        
        label {
            font-weight: bold;
            margin-bottom: 8px;
            color: #2c3e50;
        }
        
        input, select {
            padding: 10px;
            border: 2px solid #e0e0e0;
            border-radius: 5px;
            font-size: 14px;
            transition: border-color 0.3s;
        }
        
        input:focus, select:focus {
            outline: none;
            border-color: #3498db;
        }
        
        .button-group {
            display: flex;
            gap: 15px;
            justify-content: center;
            flex-wrap: wrap;
            margin-top: 20px;
        }
        
        button {
            padding: 12px 25px;
            border: none;
            border-radius: 8px;
            cursor: pointer;
            font-weight: bold;
            font-size: 14px;
            transition: all 0.3s;
            display: flex;
            align-items: center;
            gap: 8px;
        }
        
        .btn-primary {
            background: linear-gradient(135deg, #3498db, #2980b9);
            color: white;
        }
        
        .btn-success {
            background: linear-gradient(135deg, #2ecc71, #27ae60);
            color: white;
        }
        
        .btn-warning {
            background: linear-gradient(135deg, #f39c12, #e67e22);
            color: white;
        }
        
        button:hover {
            transform: translateY(-2px);
            box-shadow: 0 5px 15px rgba(0,0,0,0.2);
        }
        
        .results-panel {
            background: white;
            padding: 25px;
            border-radius: 10px;
            margin-bottom: 30px;
            border: 2px solid #e0e0e0;
        }
        
        .results-grid {
            display: grid;
            grid-template-columns: repeat(auto-fit, minmax(250px, 1fr));
            gap: 20px;
            margin-top: 20px;
        }
        
        .result-card {
            background: linear-gradient(135deg, #667eea, #764ba2);
            color: white;
            padding: 20px;
            border-radius: 10px;
            text-align: center;
        }
        
        .result-value {
            font-size: 2em;
            font-weight: bold;
            margin: 10px 0;
        }
        
        .result-label {
            font-size: 0.9em;
            opacity: 0.9;
        }
        
        .chart-container {
            background: white;
            border-radius: 10px;
            padding: 20px;
            margin-bottom: 20px;
            box-shadow: 0 5px 15px rgba(0,0,0,0.1);
        }
        
        .chart-title {
            font-size: 1.3em;
            font-weight: bold;
            margin-bottom: 15px;
            color: #2c3e50;
            text-align: center;
        }
        
        .status-message {
            padding: 15px;
            border-radius: 5px;
            margin: 10px 0;
            text-align: center;
            font-weight: bold;
        }
        
        .status-waiting {
            background: #fff3cd;
            color: #856404;
            border: 1px solid #ffeaa7;
        }
        
        .status-success {
            background: #d1ecf1;
            color: #0c5460;
            border: 1px solid #bee5eb;
        }
        
        .status-error {
            background: #f8d7da;
            color: #721c24;
            border: 1px solid #f5c6cb;
        }
        
        .progress-bar {
            width: 100%;
            height: 20px;
            background: #e0e0e0;
            border-radius: 10px;
            overflow: hidden;
            margin: 10px 0;
        }
        
        .progress {
            height: 100%;
            background: linear-gradient(135deg, #3498db, #2980b9);
            width: 0%;
            transition: width 0.3s;
        }
        
        .iteration-info {
            text-align: center;
            font-size: 1.1em;
            margin: 15px 0;
            color: #2c3e50;
        }
        
        .data-table {
            width: 100%;
            border-collapse: collapse;
            margin-top: 20px;
        }
        
        .data-table th, .data-table td {
            padding: 12px 15px;
            text-align: left;
            border-bottom: 1px solid #ddd;
        }
        
        .data-table th {
            background-color: #f8f9fa;
            font-weight: bold;
            color: #2c3e50;
        }
        
        .data-table tr:hover {
            background-color: #f5f5f5;
        }
        
        .section-title {
            font-size: 1.5em;
            font-weight: bold;
            margin: 25px 0 15px 0;
            color: #2c3e50;
            border-bottom: 2px solid #3498db;
            padding-bottom: 10px;
        }
        
        .optimization-info {
            display: grid;
            grid-template-columns: repeat(auto-fit, minmax(200px, 1fr));
            gap: 15px;
            margin-top: 20px;
        }
        
        .info-card {
            background: #f8f9fa;
            padding: 15px;
            border-radius: 8px;
            text-align: center;
            border-left: 4px solid #3498db;
        }
        
        .info-value {
            font-size: 1.5em;
            font-weight: bold;
            color: #2c3e50;
        }
        
        .info-label {
            font-size: 0.9em;
            color: #666;
        }
    </style>
</head>
<body>
    <div class="container">
        <div class="header">
            <h1>Heterojunction Composite Iterative Learning System</h1>
            <p>Material Performance Prediction and Optimization Based on Multi-Objective Optimization</p>
        </div>
        
        <div class="content">
            <!-- Input Parameters Panel -->
            <div class="input-panel">
                <h2>Material Parameter Input</h2>
                <div class="input-grid">
                    <div class="input-group">
                        <label for="valenceBand">Valence Band (eV)</label>
                        <input type="number" id="valenceBand" value="1.5" step="0.1" min="0" max="5">
                    </div>
                    <div class="input-group">
                        <label for="conductionBand">Conduction Band (eV)</label>
                        <input type="number" id="conductionBand" value="-0.5" step="0.1" min="-5" max="5">
                    </div>
                    <div class="input-group">
                        <label for="bandGap">Band Gap (eV)</label>
                        <input type="number" id="bandGap" value="2.0" step="0.1" min="0.5" max="5">
                    </div>
                    <div class="input-group">
                        <label for="absorptionRange">Light Absorption Range (nm)</label>
                        <input type="number" id="absorptionRange" value="600" step="10" min="300" max="1000">
                    </div>
                    <div class="input-group">
                        <label for="plqy">PLQY</label>
                        <input type="number" id="plqy" value="0.5" step="0.05" min="0" max="1">
                    </div>
                    <div class="input-group">
                        <label for="photocatalyticRate">Photocatalytic Rate</label>
                        <input type="number" id="photocatalyticRate" value="0.5" step="0.05" min="0" max="1">
                    </div>
                    <div class="input-group">
                        <label for="materialType">Material Type</label>
                        <select id="materialType">
                            <option value="oxide">Oxide</option>
                            <option value="sulfide">Sulfide</option>
                            <option value="nitride">Nitride</option>
                            <option value="composite">Composite</option>
                        </select>
                    </div>
                    <div class="input-group">
                        <label for="ssa">Specific Surface Area (m²/g)</label>
                        <input type="number" id="ssa" value="50" step="5" min="10" max="200">
                    </div>
                </div>
                
                <div class="button-group">
                    <button class="btn-primary" onclick="startOptimization()">
                        <span>🚀 Start Iterative Learning (500 iterations)</span>
                    </button>
                    <button class="btn-success" onclick="exportToExcel()">
                        <span>📊 Export Excel Results</span>
                    </button>
                    <button class="btn-warning" onclick="resetForm()">
                        <span>🔄 Reset</span>
                    </button>
                </div>
                
                <div id="statusMessage" class="status-message status-waiting">
                    Waiting to start iterative learning...
                </div>
                
                <div class="progress-bar">
                    <div id="optimizationProgress" class="progress"></div>
                </div>
                
                <div id="iterationInfo" class="iteration-info">
                    Iteration progress: 0 / 500
                </div>
            </div>
            
            <!-- Current Best Result -->
            <div class="results-panel">
                <h2>Current Best Result</h2>
                <div class="results-grid">
                    <div class="result-card">
                        <div class="result-label">Reward</div>
                        <div id="currentReward" class="result-value">0.000</div>
                        <div class="result-label">Overall Performance Score</div>
                    </div>
                    <div class="result-card">
                        <div class="result-label">Band Gap</div>
                        <div id="currentBandGap" class="result-value">0.000 eV</div>
                        <div class="result-label">Target: 1.8-2.2 eV</div>
                    </div>
                    <div class="result-card">
                        <div class="result-label">Light Absorption Range</div>
                        <div id="currentAbsorption" class="result-value">0 nm</div>
                        <div class="result-label">Target: 550-650 nm</div>
                    </div>
                    <div class="result-card">
                        <div class="result-label">PLQY</div>
                        <div id="currentPLQY" class="result-value">0.000</div>
                        <div class="result-label">Target: > 0.8</div>
                    </div>
                    <div class="result-card">
                        <div class="result-label">Photocatalytic Rate</div>
                        <div id="currentPhotocatalytic" class="result-value">0.000</div>
                        <div class="result-label">Target: > 0.7</div>
                    </div>
                </div>
            </div>
            
            <!-- Optimization Info -->
            <div class="optimization-info">
                <div class="info-card">
                    <div class="info-value" id="bestIteration">0</div>
                    <div class="info-label">Best Iteration</div>
                </div>
                <div class="info-card">
                    <div class="info-value" id="totalTime">0</div>
                    <div class="info-label">Total Time (s)</div>
                </div>
                <div class="info-card">
                    <div class="info-value" id="improvement">0%</div>
                    <div class="info-label">Improvement</div>
                </div>
                <div class="info-card">
                    <div class="info-value" id="paretoCount">0</div>
                    <div class="info-label">Pareto Optimal Solutions</div>
                </div>
            </div>
            
            <!-- Charts -->
            <div class="chart-container">
                <div class="chart-title">Convergence Curve of Iterative Learning</div>
                <div id="convergenceChart"></div>
            </div>
            
            <div style="display: grid; grid-template-columns: 1fr 1fr; gap: 20px;">
                <div class="chart-container">
                    <div class="chart-title">Parameter Optimization Process</div>
                    <div id="parameterChart"></div>
                </div>
                <div class="chart-container">
                    <div class="chart-title">Performance Radar Chart</div>
                    <div id="radarChart"></div>
                </div>
            </div>
            
            <div class="chart-container">
                <div class="chart-title">Pareto Front Analysis</div>
                <div id="paretoChart"></div>
            </div>
            
            <!-- Results Table -->
            <div class="section-title">Iterative Learning Results</div>
            <div class="chart-container">
                <div class="chart-title">Top 20 Best Results</div>
                <table class="data-table" id="resultsTable">
                    <thead>
                        <tr>
                            <th>Iteration</th>
                            <th>Reward</th>
                            <th>Band Gap (eV)</th>
                            <th>Absorption (nm)</th>
                            <th>PLQY</th>
                            <th>Photocatalytic Rate</th>
                            <th>Valence Band (eV)</th>
                            <th>Conduction Band (eV)</th>
                        </tr>
                    </thead>
                    <tbody id="resultsBody">
                        <!-- Results will be added dynamically via JavaScript -->
                    </tbody>
                </table>
            </div>
        </div>
    </div>

    <script>
        // Heterojunction Composite Optimizer
        class HeterojunctionOptimizer {
            constructor() {
                this.iterationCount = 500;
                this.history = [];
                this.bestSolution = null;
                this.isRunning = false;
                this.paretoFront = [];
                this.startTime = null;
            }
            
            // Calculate reward function
            calculateReward(params) {
                const {
                    valenceBand,
                    conductionBand,
                    bandGap,
                    absorptionRange,
                    plqy,
                    photocatalyticRate,
                    materialType,
                    ssa
                } = params;
                
                // Band gap score - ideal range 1.8-2.2 eV
                let bandGapScore = 0;
                if (bandGap >= 1.8 && bandGap <= 2.2) {
                    bandGapScore = 1 - Math.abs(bandGap - 2.0) / 0.4;
                } else {
                    bandGapScore = Math.max(0, 1 - Math.abs(bandGap - 2.0) / 1.0);
                }
                
                // Light absorption score - ideal range 550-650 nm
                let absorptionScore = 0;
                if (absorptionRange >= 550 && absorptionRange <= 650) {
                    absorptionScore = 1 - Math.abs(absorptionRange - 600) / 100;
                } else {
                    absorptionScore = Math.max(0, 1 - Math.abs(absorptionRange - 600) / 300);
                }
                
                // PLQY score - higher is better
                const plqyScore = Math.min(plqy / 0.9, 1);
                
                // Photocatalytic rate score - higher is better
                const photocatalyticScore = Math.min(photocatalyticRate / 0.8, 1);
                
                // Band alignment score - important for heterojunction
                const bandAlignmentScore = this.calculateBandAlignmentScore(valenceBand, conductionBand);
                
                // Material type weight
                const materialWeights = {
                    'oxide': 1.0,
                    'sulfide': 1.1,
                    'nitride': 1.2,
                    'composite': 1.3
                };
                const materialWeight = materialWeights[materialType] || 1.0;
                
                // Specific surface area score
                const ssaScore = Math.min(ssa / 100, 1);
                
                // Comprehensive reward
                const reward = (
                    bandGapScore * 0.25 +
                    absorptionScore * 0.20 +
                    plqyScore * 0.20 +
                    photocatalyticScore * 0.15 +
                    bandAlignmentScore * 0.10 +
                    ssaScore * 0.10
                ) * materialWeight;
                
                return {
                    reward,
                    bandGapScore,
                    absorptionScore,
                    plqyScore,
                    photocatalyticScore,
                    bandAlignmentScore,
                    ssaScore
                };
            }
            
            // Calculate band alignment score
            calculateBandAlignmentScore(vb, cb) {
                // Ideal heterojunction band alignment: type II alignment facilitates charge separation
                // Simplified calculation here; actual evaluation should be based on specific material pairs
                const vbScore = 1 - Math.abs(vb - 1.5) / 2.0;
                const cbScore = 1 - Math.abs(cb + 0.5) / 1.5;
                return (vbScore + cbScore) / 2;
            }
            
            // Generate random parameter mutation
            generateMutation(baseParams, iteration) {
                const mutationRate = Math.max(0.1, 0.5 - iteration / 1000);
                
                return {
                    valenceBand: this.mutateValue(baseParams.valenceBand, 0.2, mutationRate, 0, 5),
                    conductionBand: this.mutateValue(baseParams.conductionBand, 0.2, mutationRate, -5, 0),
                    bandGap: this.mutateValue(baseParams.bandGap, 0.1, mutationRate, 0.5, 5),
                    absorptionRange: this.mutateValue(baseParams.absorptionRange, 20, mutationRate, 300, 1000),
                    plqy: this.mutateValue(baseParams.plqy, 0.05, mutationRate, 0, 1),
                    photocatalyticRate: this.mutateValue(baseParams.photocatalyticRate, 0.05, mutationRate, 0, 1),
                    materialType: baseParams.materialType, // keep unchanged for now
                    ssa: this.mutateValue(baseParams.ssa, 10, mutationRate, 10, 200)
                };
            }
            
            // Numerical mutation function
            mutateValue(value, range, mutationRate, min, max) {
                if (Math.random() < mutationRate) {
                    const newValue = value + (Math.random() - 0.5) * range * 2;
                    return Math.max(min, Math.min(max, newValue));
                }
                return value;
            }
            
            // Execute iterative learning
            async optimize(initialParams, onProgress) {
                this.isRunning = true;
                this.history = [];
                this.paretoFront = [];
                this.startTime = Date.now();
                
                let currentParams = {...initialParams};
                let bestReward = 0;
                
                for (let i = 0; i < this.iterationCount && this.isRunning; i++) {
                    // Generate mutated parameters
                    const mutatedParams = this.generateMutation(currentParams, i);
                    
                    // Calculate reward
                    const rewardResult = this.calculateReward(mutatedParams);
                    const totalReward = rewardResult.reward;
                    
                    // Record history
                    const iterationResult = {
                        iteration: i + 1,
                        params: {...mutatedParams},
                        reward: totalReward,
                        ...rewardResult
                    };
                    
                    this.history.push(iterationResult);
                    
                    // Update best solution
                    if (totalReward > bestReward) {
                        bestReward = totalReward;
                        currentParams = {...mutatedParams};
                        this.bestSolution = iterationResult;
                    }
                    
                    // Update Pareto front
                    this.updateParetoFront(iterationResult);
                    
                    // Progress callback
                    if (onProgress) {
                        onProgress(i + 1, this.iterationCount, this.bestSolution);
                    }
                    
                    // Yield control for UI update
                    await new Promise(resolve => setTimeout(resolve, 10));
                }
                
                this.isRunning = false;
                return this.bestSolution;
            }
            
            // Update Pareto front
            updateParetoFront(solution) {
                // Simplified Pareto front calculation based on reward and key performance indicators
                const isDominated = this.paretoFront.some(front => 
                    front.reward >= solution.reward &&
                    front.bandGapScore >= solution.bandGapScore &&
                    front.plqyScore >= solution.plqyScore &&
                    front.photocatalyticScore >= solution.photocatalyticScore
                );
                
                if (!isDominated) {
                    // Remove solutions dominated by the new solution
                    this.paretoFront = this.paretoFront.filter(front => 
                        !(solution.reward >= front.reward &&
                          solution.bandGapScore >= front.bandGapScore &&
                          solution.plqyScore >= front.plqyScore &&
                          solution.photocatalyticScore >= front.photocatalyticScore)
                    );
                    
                    this.paretoFront.push(solution);
                    
                    // Sort by reward
                    this.paretoFront.sort((a, b) => b.reward - a.reward);
                    
                    // Limit Pareto front size
                    if (this.paretoFront.length > 20) {
                        this.paretoFront = this.paretoFront.slice(0, 20);
                    }
                }
            }
            
            stop() {
                this.isRunning = false;
            }
            
            // Get results data for export
            getResultsData() {
                return this.history.map(record => ({
                    'Iteration': record.iteration,
                    'Reward': record.reward.toFixed(4),
                    'Band Gap (eV)': record.params.bandGap.toFixed(3),
                    'Absorption Range (nm)': record.params.absorptionRange.toFixed(1),
                    'PLQY': record.params.plqy.toFixed(4),
                    'Photocatalytic Rate': record.params.photocatalyticRate.toFixed(4),
                    'Valence Band (eV)': record.params.valenceBand.toFixed(3),
                    'Conduction Band (eV)': record.params.conductionBand.toFixed(3),
                    'Material Type': record.params.materialType,
                    'Specific Surface Area (m²/g)': record.params.ssa.toFixed(1),
                    'Band Gap Score': record.bandGapScore.toFixed(4),
                    'Absorption Score': record.absorptionScore.toFixed(4),
                    'PLQY Score': record.plqyScore.toFixed(4),
                    'Photocatalytic Score': record.photocatalyticScore.toFixed(4),
                    'Band Alignment Score': record.bandAlignmentScore.toFixed(4),
                    'Specific Surface Area Score': record.ssaScore.toFixed(4)
                }));
            }
        }

        // Global optimizer instance
        const optimizer = new HeterojunctionOptimizer();
        
        // Get input parameters
        function getInputParams() {
            return {
                valenceBand: parseFloat(document.getElementById('valenceBand').value),
                conductionBand: parseFloat(document.getElementById('conductionBand').value),
                bandGap: parseFloat(document.getElementById('bandGap').value),
                absorptionRange: parseFloat(document.getElementById('absorptionRange').value),
                plqy: parseFloat(document.getElementById('plqy').value),
                photocatalyticRate: parseFloat(document.getElementById('photocatalyticRate').value),
                materialType: document.getElementById('materialType').value,
                ssa: parseFloat(document.getElementById('ssa').value)
            };
        }
        
        // Update current best result display
        function updateCurrentResults(results) {
            if (!results) return;
            
            document.getElementById('currentReward').textContent = results.reward.toFixed(4);
            document.getElementById('currentBandGap').textContent = results.params.bandGap.toFixed(3) + ' eV';
            document.getElementById('currentAbsorption').textContent = results.params.absorptionRange.toFixed(1) + ' nm';
            document.getElementById('currentPLQY').textContent = results.params.plqy.toFixed(4);
            document.getElementById('currentPhotocatalytic').textContent = results.params.photocatalyticRate.toFixed(4);
        }
        
        // Update progress
        function updateProgress(current, total, results) {
            const progress = (current / total) * 100;
            document.getElementById('optimizationProgress').style.width = progress + '%';
            document.getElementById('iterationInfo').textContent = 
                `Iteration progress: ${current} / ${total}`;
            
            // Update elapsed time
            const elapsedTime = ((Date.now() - optimizer.startTime) / 1000).toFixed(1);
            document.getElementById('totalTime').textContent = elapsedTime;
            
            if (results) {
                updateCurrentResults(results);
                document.getElementById('bestIteration').textContent = results.iteration;
                
                // Calculate improvement
                if (current > 1) {
                    const initialReward = optimizer.history[0].reward;
                    const improvement = ((results.reward - initialReward) / initialReward * 100).toFixed(1);
                    document.getElementById('improvement').textContent = improvement + '%';
                }
                
                document.getElementById('paretoCount').textContent = optimizer.paretoFront.length;
                
                // Update results table
                updateResultsTable();
            }
        }
        
        // Start optimization
        async function startOptimization() {
            const initialParams = getInputParams();
            
            document.getElementById('statusMessage').textContent = 'Iterative learning in progress...';
            document.getElementById('statusMessage').className = 'status-message status-success';
            
            try {
                const results = await optimizer.optimize(initialParams, updateProgress);
                
                document.getElementById('statusMessage').textContent = 
                    `Iterative learning completed! Best Reward: ${results.reward.toFixed(4)} (Iteration ${results.iteration})`;
                
                // Update all charts
                updateAllCharts();
                
            } catch (error) {
                document.getElementById('statusMessage').textContent = 'Optimization error: ' + error.message;
                document.getElementById('statusMessage').className = 'status-message status-error';
            }
        }
        
        // Export to Excel
        function exportToExcel() {
            if (!optimizer.bestSolution) {
                alert('Please complete iterative learning first!');
                return;
            }
            
            const data = optimizer.getResultsData();
            const worksheet = XLSX.utils.json_to_sheet(data);
            const workbook = XLSX.utils.book_new();
            XLSX.utils.book_append_sheet(workbook, worksheet, 'Iterative Learning Results');
            
            // Generate filename
            const timestamp = new Date().toISOString().slice(0, 19).replace(/:/g, '-');
            const filename = `Heterojunction_Iterative_Learning_${timestamp}.xlsx`;
            
            // Export file
            XLSX.writeFile(workbook, filename);
        }
        
        // Reset form
        function resetForm() {
            optimizer.stop();
            optimizer.history = [];
            optimizer.bestSolution = null;
            optimizer.paretoFront = [];
            
            document.getElementById('optimizationProgress').style.width = '0%';
            document.getElementById('iterationInfo').textContent = 'Iteration progress: 0 / 500';
            document.getElementById('statusMessage').textContent = 'Optimizer reset.';
            document.getElementById('statusMessage').className = 'status-message status-waiting';
            
            // Reset result display
            updateCurrentResults({
                reward: 0,
                params: {
                    bandGap: 0,
                    absorptionRange: 0,
                    plqy: 0,
                    photocatalyticRate: 0
                }
            });
            
            // Reset info cards
            document.getElementById('bestIteration').textContent = '0';
            document.getElementById('totalTime').textContent = '0';
            document.getElementById('improvement').textContent = '0%';
            document.getElementById('paretoCount').textContent = '0';
            
            // Clear table
            document.getElementById('resultsBody').innerHTML = '';
        }
        
        // Update results table
        function updateResultsTable() {
            const tbody = document.getElementById('resultsBody');
            tbody.innerHTML = '';
            
            // Display top 20 best results
            const topResults = [...optimizer.history]
                .sort((a, b) => b.reward - a.reward)
                .slice(0, 20);
            
            topResults.forEach(result => {
                const row = document.createElement('tr');
                row.innerHTML = `
                    <td>${result.iteration}</td>
                    <td>${result.reward.toFixed(4)}</td>
                    <td>${result.params.bandGap.toFixed(3)}</td>
                    <td>${result.params.absorptionRange.toFixed(1)}</td>
                    <td>${result.params.plqy.toFixed(4)}</td>
                    <td>${result.params.photocatalyticRate.toFixed(4)}</td>
                    <td>${result.params.valenceBand.toFixed(3)}</td>
                    <td>${result.params.conductionBand.toFixed(3)}</td>
                `;
                tbody.appendChild(row);
            });
        }
        
        // Update all charts
        function updateAllCharts() {
            updateConvergenceChart();
            updateParameterChart();
            updateRadarChart();
            updateParetoChart();
        }
        
        // Convergence curve chart
        function updateConvergenceChart() {
            if (optimizer.history.length === 0) return;
            
            const iterations = optimizer.history.map(h => h.iteration);
            const rewards = optimizer.history.map(h => h.reward);
            
            const trace = {
                x: iterations,
                y: rewards,
                type: 'scatter',
                mode: 'lines',
                name: 'Reward',
                line: { color: '#3498db', width: 2 }
            };
            
            const layout = {
                title: 'Convergence Curve of Iterative Learning',
                xaxis: { title: 'Iteration' },
                yaxis: { title: 'Reward' }
            };
            
            Plotly.newPlot('convergenceChart', [trace], layout);
        }
        
        // Parameter optimization process chart
        function updateParameterChart() {
            if (optimizer.history.length === 0) return;
            
            const iterations = optimizer.history.map(h => h.iteration);
            
            const traces = [
                {
                    x: iterations,
                    y: optimizer.history.map(h => h.params.bandGap),
                    name: 'Band Gap',
                    line: { color: '#e74c3c' }
                },
                {
                    x: iterations,
                    y: optimizer.history.map(h => h.params.absorptionRange),
                    name: 'Absorption',
                    yaxis: 'y2',
                    line: { color: '#2ecc71' }
                },
                {
                    x: iterations,
                    y: optimizer.history.map(h => h.params.plqy),
                    name: 'PLQY',
                    yaxis: 'y3',
                    line: { color: '#9b59b6' }
                },
                {
                    x: iterations,
                    y: optimizer.history.map(h => h.params.photocatalyticRate),
                    name: 'Photocatalytic Rate',
                    yaxis: 'y4',
                    line: { color: '#f39c12' }
                }
            ];
            
            const layout = {
                title: 'Parameter Optimization Process',
                xaxis: { title: 'Iteration' },
                yaxis: { title: 'Band Gap (eV)', side: 'left' },
                yaxis2: { title: 'Absorption (nm)', overlaying: 'y', side: 'right' },
                yaxis3: { title: 'PLQY', overlaying: 'y', side: 'right', position: 0.85 },
                yaxis4: { title: 'Photocatalytic Rate', overlaying: 'y', side: 'right', position: 0.95 }
            };
            
            Plotly.newPlot('parameterChart', traces, layout);
        }
        
        // Radar chart
        function updateRadarChart() {
            if (!optimizer.bestSolution) return;
            
            const categories = ['Band Gap Score', 'Absorption Score', 'PLQY Score', 'Photocatalytic Score', 'Band Alignment Score', 'Specific Surface Area Score'];
            const bestValues = [
                optimizer.bestSolution.bandGapScore,
                optimizer.bestSolution.absorptionScore,
                optimizer.bestSolution.plqyScore,
                optimizer.bestSolution.photocatalyticScore,
                optimizer.bestSolution.bandAlignmentScore,
                optimizer.bestSolution.ssaScore
            ];
            
            // Close radar chart
            const bestRadarValues = [...bestValues, bestValues[0]];
            const angles = [...Array(categories.length).keys()].map(i => 
                i * 2 * Math.PI / categories.length
            );
            angles.push(angles[0]);
            
            const trace = {
                r: bestRadarValues,
                theta: angles.map(angle => angle * 180 / Math.PI),
                fill: 'toself',
                name: 'Best Parameters',
                line: { color: '#3498db' },
                fillcolor: 'rgba(52, 152, 219, 0.3)'
            };
            
            const layout = {
                polar: {
                    radialaxis: {
                        visible: true,
                        range: [0, 1]
                    }
                },
                showlegend: false
            };
            
            Plotly.newPlot('radarChart', [trace], layout);
        }
        
        // Pareto front chart
        function updateParetoChart() {
            if (optimizer.paretoFront.length === 0) return;
            
            const trace = {
                x: optimizer.paretoFront.map(sol => sol.params.bandGap),
                y: optimizer.paretoFront.map(sol => sol.reward),
                mode: 'markers',
                type: 'scatter',
                name: 'Pareto Front',
                marker: {
                    size: 10,
                    color: optimizer.paretoFront.map(sol => sol.params.plqy),
                    colorscale: 'Viridis',
                    showscale: true,
                    colorbar: { title: 'PLQY' }
                },
                text: optimizer.paretoFront.map(sol => 
                    `Iteration: ${sol.iteration}<br>Reward: ${sol.reward.toFixed(4)}<br>Band Gap: ${sol.params.bandGap.toFixed(3)} eV<br>PLQY: ${sol.params.plqy.toFixed(4)}`
                ),
                hoverinfo: 'text'
            };
            
            const layout = {
                title: 'Pareto Front - Band Gap vs Reward (Color indicates PLQY)',
                xaxis: { title: 'Band Gap (eV)' },
                yaxis: { title: 'Reward' }
            };
            
            Plotly.newPlot('paretoChart', [trace], layout);
        }
        
        // Initialize page
        function initializePage() {
            updateCurrentResults({
                reward: 0,
                params: {
                    bandGap: 0,
                    absorptionRange: 0,
                    plqy: 0,
                    photocatalyticRate: 0
                }
            });
            
            // Initialize charts
            updateAllCharts();
        }
        
        // Initialize after page loads
        window.addEventListener('DOMContentLoaded', initializePage);
    </script>
</body>
</html>
'''

HTML(html_content)
```
